```python
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta

# =========================
# CONFIGURATION
# =========================
YEARS = 5 # Changed to 5 years to match the CSV data range
END_DATE = datetime(2018, 2, 7) # Set to the last date available in the CSV
START_DATE = END_DATE - timedelta(days=int(YEARS * 365))

US_TICKERS = [
    "AAPL", "MSFT", "AMZN", "GOOGL", "META",
    "TSLA", "NVDA", "NFLX", "JPM", "AMD"
]

INDIA_TICKERS = [
    "RELIANCE.NS", "TCS.NS", "INFY.NS",
    "HDFCBANK.NS", "ICICIBANK.NS",
    "SBIN.NS", "ITC.NS",
    "LT.NS", "WIPRO.NS", "BHARTIARTL.NS"
]

# =========================
# LOAD US DATA (From CSV)
# =========================
def load_us_data(csv_path):
    df = pd.read_csv("/content/all_stocks_5yr.csv")

    # Ensure 'date' column exists, handling 'Date' if present
    if 'Date' in df.columns:
        df = df.rename(columns={'Date': 'date'})
    elif 'date' not in df.columns:
        print("Error: Neither 'Date' nor 'date' column found in CSV.")
        return pd.DataFrame() # Return empty DataFrame if no date column

    df['date'] = pd.to_datetime(df['date'])

    # Debugging prints
    print(f"--- Debugging load_us_data ---")
    print(f"Initial rows after date conversion: {len(df)}")
    print(f"Unique 'Name' values in raw data: {df['Name'].unique().tolist()}")
    print(f"Date range in raw data: {df['date'].min()} to {df['date'].max()}")
    print(f"Configured START_DATE: {START_DATE}")

    # Filter by US_TICKERS
    initial_rows = len(df)
    df = df[df['Name'].isin(US_TICKERS)]
    print(f"Rows after filtering by US_TICKERS ({len(US_TICKERS)} tickers): {len(df)}")
    if len(df) == 0:
        print("Warning: No data remaining after filtering by US_TICKERS. Check if ticker names in CSV match US_TICKERS.")
        # Identify which tickers are actually present in the dataframe
        all_tickers_in_csv = set(df['Name'].unique().tolist())
        present_tickers = all_tickers_in_csv.intersection(set(US_TICKERS))
        print(f"Tickers found in CSV that are also in US_TICKERS: {list(present_tickers)}")


    # Filter by date range
    initial_rows = len(df)
    df = df[df['date'] >= START_DATE]
    print(f"Rows after filtering by START_DATE ({START_DATE}): {len(df)}")
    if len(df) == 0 and initial_rows > 0: # Only warn if there was data before date filter
        print("Warning: No data remaining after filtering by START_DATE. Check date ranges.")

    print(f"Final DF shape in load_us_data: {df.shape}")
    print(f"--- End Debugging load_us_data ---")

    return df

# =========================
# LOAD INDIAN DATA (Yahoo)
# =========================
def load_india_data():
    data = yf.download(
        INDIA_TICKERS,
        start=START_DATE,
        end=END_DATE,
        group_by='ticker'
    )
    return data
```

```python
!pip install ta
import pandas as pd
import numpy as np
from ta.momentum import RSIIndicator
from ta.trend import MACD
from ta.trend import SMAIndicator
from sklearn.preprocessing import MinMaxScaler


# =========================
# ADD TECHNICAL INDICATORS
# =========================
def add_indicators(df):

    df = df.copy()

    df['rsi'] = RSIIndicator(close=df['close'], window=14).rsi()
    df['sma'] = SMAIndicator(close=df['close'], window=14).sma_indicator()

    macd = MACD(close=df['close'])
    df['macd'] = macd.macd()

    df.dropna(inplace=True)

    return df


# =========================
# PROCESS MULTI-STOCK DATA
# =========================
def process_portfolio_data(df):

    stocks = df['Name'].unique()
    processed = []

    for stock in stocks:
        temp = df[df['Name'] == stock].sort_values('date')
        temp = add_indicators(temp)
        temp['Name'] = stock
        processed.append(temp)

    final_df = pd.concat(processed)
    return final_df


# =========================
# NORMALIZATION
# =========================
def normalize_features(df):

    scaler = MinMaxScaler()

    feature_cols = ['close', 'volume', 'rsi', 'sma', 'macd']
    df[feature_cols] = scaler.fit_transform(df[feature_cols])

    return df, scaler


# =========================
# TRAIN TEST SPLIT (70-30)
# =========================
def train_test_split(df):

    unique_dates = sorted(df['date'].unique())
    split_idx = int(len(unique_dates) * 0.7)

    train_dates = unique_dates[:split_idx]
    test_dates = unique_dates[split_idx:]

    train_df = df[df['date'].isin(train_dates)]
    test_df = df[df['date'].isin(test_dates)]

    return train_df, test_df
```

```python
import numpy as np

class PortfolioEnv:

    def __init__(self, df, initial_capital=100000, transaction_cost=0.001, risk_lambda=0.1):

        self.df = df
        self.initial_capital = initial_capital
        self.transaction_cost = transaction_cost
        self.risk_lambda = risk_lambda

        self.stocks = sorted(df['Name'].unique())
        self.dates = sorted(df['date'].unique())

        self.n_stocks = len(self.stocks)
        self.reset()

    # =========================
    # RESET ENVIRONMENT
    # =========================
    def reset(self):

        self.current_step = 0
        self.portfolio_value = self.initial_capital
        self.weights = np.ones(self.n_stocks) / self.n_stocks
        self.returns_history = []

        return self._get_state()

    # =========================
    # SOFTMAX FUNCTION
    # =========================
    def softmax(self, x):
        exp_x = np.exp(x - np.max(x))
        return exp_x / exp_x.sum()

    # =========================
    # GET STATE
    # =========================
    def _get_state(self):

        date = self.dates[self.current_step]
        daily_data = self.df[self.df['date'] == date]

        state = []

        for stock in self.stocks:
            row = daily_data[daily_data['Name'] == stock]
            features = row[['close', 'volume', 'rsi', 'sma', 'macd']].values.flatten()
            state.extend(features)

        state.append(self.portfolio_value / self.initial_capital)

        return np.array(state, dtype=np.float32)

    # =========================
    # STEP FUNCTION
    # =========================
    def step(self, action):

        done = False

        # Convert action to softmax weights
        new_weights = self.softmax(action)

        prev_value = self.portfolio_value

        # Get today's and next day's prices
        date_today = self.dates[self.current_step]
        date_next = self.dates[self.current_step + 1]

        today_data = self.df[self.df['date'] == date_today]
        next_data = self.df[self.df['date'] == date_next]

        daily_returns = []

        for i, stock in enumerate(self.stocks):

            price_today = today_data[today_data['Name'] == stock]['close'].values[0]
            price_next = next_data[next_data['Name'] == stock]['close'].values[0]

            stock_return = (price_next - price_today) / price_today
            daily_returns.append(stock_return)

        daily_returns = np.array(daily_returns)

        # Portfolio return
        portfolio_return = np.dot(new_weights, daily_returns)

        # Transaction cost
        turnover = np.sum(np.abs(new_weights - self.weights))
        cost = self.transaction_cost * turnover

        portfolio_return -= cost

        # Update portfolio value
        self.portfolio_value *= (1 + portfolio_return)

        self.weights = new_weights
        self.returns_history.append(portfolio_return)

        # Risk penalty (Sharpe-style)
        if len(self.returns_history) > 30:
            volatility = np.std(self.returns_history[-30:])
        else:
            volatility = np.std(self.returns_history)

        reward = portfolio_return - self.risk_lambda * volatility

        self.current_step += 1

        if self.current_step >= len(self.dates) - 1:
            done = True

        next_state = self._get_state()

        return next_state, reward, done, {}
```

```python
import numpy as np
import random
from collections import deque
import tensorflow as tf
from tensorflow.keras import models, layers, optimizers

class DQNAgent:

    def __init__(self, state_size, action_size):

        self.state_size = state_size
        self.action_size = action_size

        self.memory = deque(maxlen=5000)

        self.gamma = 0.95
        self.epsilon = 1.0
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.learning_rate = 0.001
        self.batch_size = 64

        self.model = self._build_model()
        self.target_model = self._build_model()
        self.update_target_model()

    # =========================
    # BUILD NETWORK
    # =========================
    def _build_model(self):

        model = models.Sequential()
        model.add(layers.Dense(128, activation='relu', input_dim=self.state_size))
        model.add(layers.Dense(64, activation='relu'))
        model.add(layers.Dense(32, activation='relu'))
        model.add(layers.Dense(self.action_size, activation='linear'))

        model.compile(
            loss='mse',
            optimizer=optimizers.Adam(learning_rate=self.learning_rate)
        )

        return model

    # =========================
    # TARGET UPDATE
    # =========================
    def update_target_model(self):
        self.target_model.set_weights(self.model.get_weights())

    # =========================
    # STORE EXPERIENCE
    # =========================
    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    # =========================
    # ACTION SELECTION
    # =========================
    def act(self, state):

        if np.random.rand() <= self.epsilon:
            return np.random.randn(self.action_size)

        q_values = self.model.predict(state[np.newaxis], verbose=0)[0]
        return q_values

    # =========================
    # TRAINING STEP
    # =========================
    def replay(self):

        if len(self.memory) < self.batch_size:
            return

        minibatch = random.sample(self.memory, self.batch_size)

        states = []
        targets = []

        for state, action, reward, next_state, done in minibatch:

            target = self.model.predict(state[np.newaxis], verbose=0)[0]

            if done:
                target = reward
            else:
                t = self.target_model.predict(next_state[np.newaxis], verbose=0)[0]
                target = reward + self.gamma * np.max(t)

            states.append(state)
            targets.append(np.full(self.action_size, target))

        self.model.fit(
            np.array(states),
            np.array(targets),
            epochs=1,
            verbose=0
        )

        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay
```

```python
import numpy as np

def train_dqn(train_df, episodes=50):

    env = PortfolioEnv(train_df)
    state_size = len(env.reset())
    action_size = env.n_stocks

    agent = DQNAgent(state_size, action_size)

    for e in range(episodes):

        state = env.reset()
        total_reward = 0

        while True:

            action = agent.act(state)
            next_state, reward, done, _ = env.step(action)

            agent.remember(state, action, reward, next_state, done)
            agent.replay()

            state = next_state
            total_reward += reward

            if done:
                break

        agent.update_target_model()

        print(f"Episode {e+1}/{episodes} | Reward: {total_reward:.4f}")

    agent.model.save("dqn_portfolio_model.h5")
    return agent
```
